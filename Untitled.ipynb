```python
import torch 
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer(return_X_y=True)

X = torch.tensor(data[0], dtype=torch.float32)
y = torch.tensor(data[1], dtype=torch.float32).unsqueeze(-1)
```

```python
n = int(len(X) * .7)

X_train = X[:n]
X_test = X[n:]
y_train = y[:n]
y_test = y[n:]

print(X_train.shape); print(X_test.shape)
```

```python
class MLP(nn.Module):
    def __init__(self, n_features):
        super().__init__()
        self.fc1 = nn.Linear(n_features, 21)
        self.fc2 = nn.Linear(21, 18)
        self.fc3 = nn.Linear(18, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, X):
        x = self.fc1(X)
        x = self.fc2(x)
        out = self.fc3(x)
        return out
```

```python
n_features = X_train.shape[1]

model = MLP(n_features)
# defining loss and optimizers
critation = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

for k in range(500):
    y_pred = model(X_train)
    loss = critation(y_pred, y_train)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (k+1) % 50 == 0:
        print(f'Iter: {k+1} loss={loss.item()}')
```

```python
testing_data = torch.tensor(X_test[0])

y_pred = model(testing_data)
```
