```python
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
# f(x) = (x - 3)^2

x = torch.tensor(2.0, requires_grad=True)
y = (x - 3) ** 2
y.backward()

print(x.grad)
```

```python
# f(x, y) =  (xy-1)^2
x = torch.tensor(2.0, requires_grad=True)
y = torch.tensor(3.0, requires_grad=True)

f = (x*y - 1)**2

f.backward()


print(x.grad)
print(y.grad)
```

```python
# h(x) = x^1/4
x = torch.tensor(16.0, requires_grad=True)
h = (x)**0.25

h.backward()

print(x.grad)
```

```python
# h(x) = x^-1
x = torch.tensor(-1.0, requires_grad=True)
h = (x)**-7

h.backward()

print(x.grad)
```
